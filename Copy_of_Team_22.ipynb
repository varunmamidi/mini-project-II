importing Dependencies



```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

uploading the dataset to a pandas Dataframe

```python
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

printing the first five rows of the dataset

```python
credit_card_data.head()
```

printing the last five rows of the dataset

```python
credit_card_data.tail()
```

information about the dataset

```python
credit_card_data.info()
```

checking the number of missing values in the each column

```python
credit_card_data.isnull().sum()
```

distribution of the legit transactioin & fradulent transcation

```python
credit_card_data['Class'].value_counts()
```

This dataset is highly unbalanced

0 ---> Normal Transaction
1 ---> Fradulant Transaction

seperating the Data for analysis

```python
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

Statistical Methods of the data

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

Compare the values for both Transactions

```python
credit_card_data.groupby('Class').mean()
```

Under-sampling

Build a sample containg similar distribution of normal transaction and Fradulant Transactions

number of fradulent transcaions --> 492

```python
legit_sample = legit.sample(n=492)
```

concatinating two data frames

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into Features & Targets

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split the data into Training data & Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

training the Logistic Regression Model with Training Data

```python
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

accuracy on training data

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

accuracy on test data

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
